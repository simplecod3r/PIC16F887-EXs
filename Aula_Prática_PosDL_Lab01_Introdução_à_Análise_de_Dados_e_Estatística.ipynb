## **Probabilidade, Distribuições Discretas e Contínuas**

Conceitos:

* Evento, espaço amostral, probabilidade de um evento.

* Probabilidade condicional e independência.

Probabilidade mede a chance de um evento ocorrer dentro de um espaço amostral.

Aqui vamos simular o lançamento de um dado e calcular probabilidades simples e condicionais.

```python
import numpy as np

# Simulação de um lançamento de dado
dado = np.random.randint(1, 7, size=10000)
prob_6 = np.mean(dado == 6)
prob_par = np.mean(dado % 2 == 0)
prob_6_dado_par = np.mean((dado == 6) & (dado % 2 == 0)) / prob_par

print(f"P(6) = {prob_6:.3f}")
print(f"P(Par) = {prob_par:.3f}")
print(f"P(6 | Par) = {prob_6_dado_par:.3f}")
```

Explicação:

- Estamos gerando 10.000 lançamentos para estimar probabilidades de forma empírica.

- np.mean(condição) conta quantas vezes a condição é verdadeira e divide pelo total.

- P(6) é a chance de sair 6, P(Par) é a chance de sair número par.

- P(6 | Par) é probabilidade condicional, mostra que, sabendo que o número é par, a chance de ser 6 é 1/3.

### **Distribuições Discretas**

* Bernoulli

Modela um evento binário (sucesso/fracasso).

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

p = 0.3
x = [0, 1]
pmf = bernoulli.pmf(x, p)

plt.bar(x, pmf)
plt.title('Distribuição Bernoulli (p=0.3)')
plt.xticks(x, ['Fracasso (0)', 'Sucesso (1)'])
plt.ylabel('Probabilidade')
plt.show()
```

Explicação:

- A Bernoulli representa um experimento com dois resultados possíveis: sucesso (1) ou fracasso (0).

- p=0.3 significa que a chance de sucesso é 30%.

- bernoulli.pmf(x, p) calcula a probabilidade de cada valor possível (0 ou 1).

- O gráfico mostra uma barra em 0 e outra em 1, refletindo suas probabilidades.

* Binomial

Exemplo: 10 lançamentos de moeda.

```python
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

plt.bar(x, pmf)
plt.title('Distribuição Binomial (n=10, p=0.5)')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.show()
```

Explicação:

- A Binomial modela o número de sucessos em n tentativas independentes.

- Aqui, simulamos 10 lançamentos de moeda (sucesso = “cara”).

- pmf = probabilidade de cada quantidade possível de sucessos (0 a 10).

- cdf = probabilidade acumulada até certo ponto.

* Poisson

Exemplo: número de e-mails recebidos por hora.

```python
from scipy.stats import poisson

lambda_ = 5  # média de 5 eventos/hora
x = np.arange(0, 15)
pmf = poisson.pmf(x, lambda_)

plt.bar(x, pmf)
plt.title('Distribuição de Poisson (λ=5)')
plt.xlabel('Número de eventos')
plt.ylabel('Probabilidade')
plt.show()
```

Explicação:

- A Poisson descreve o número de vezes que um evento acontece em um intervalo fixo de tempo.

- λ = 5 significa que, em média, 5 eventos ocorrem por hora.

- O gráfico mostra a chance de ocorrer 0, 1, 2… até 15 eventos.

### **Distribuições Contínuas**

* Normal

Exemplo: alturas humanas ou médias de amostras.

```python
from scipy.stats import norm

# Mostrar diferença entre PDF (densidade) e CDF (acumulada).

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)
cdf = norm.cdf(x, loc=0, scale=1)

plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF', linestyle='--')
plt.title('Distribuição Normal Padrão')
plt.legend()
plt.show()
```

Explicação:

- A Normal é a famosa “curva em forma de sino”.

- pdf = função de densidade de probabilidade (altura da curva).

- cdf = probabilidade acumulada até certo ponto.

* Gama / Exponencial

Modela tempo até um evento (ex: tempo até falha).

```python
from scipy.stats import expon

x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=2)
plt.plot(x, pdf)
plt.title('Distribuição Exponencial (scale=2)')
plt.show()
```

Explicação:

- Modela o tempo até um evento ocorrer — como o tempo entre chamadas ou falhas.

- scale=2 significa que o tempo médio entre eventos é 2 unidades.


Dado isso: Quando os eventos acontecem de forma aleatória, mas com uma média constante de tempo entre eles, a exponencial descreve bem esse comportamento.

## **Distribuições Amostrais e Estimação**


Distribuição amostral é a distribuição de uma estatística (como média, proporção, variância etc.) calculada a partir de amostras aleatórias de uma mesma população.
Ela mostra como essa estatística varia de amostra para amostra.

- Exemplo:

Imagine que você tem uma população de flores (como no dataset Iris 🌸).
Se você tirar várias amostras diferentes e calcular a média da largura da sépala em cada uma, os valores não serão idênticos, formarão uma distribuição própria: a distribuição amostral da média.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris(as_frame=True)
df = iris.frame

# Selecionar uma variável numérica
dados = df['sepal width (cm)']

# Parâmetros da população
media_pop = dados.mean()
desvio_pop = dados.std()

print(f"Média populacional: {media_pop:.2f}")
print(f"Desvio padrão populacional: {desvio_pop:.2f}")
```

Aqui usamos todo o conjunto Iris como se fosse a “população”. Calculamos sua média e desvio padrão, ou seja, os parâmetros verdadeiros que queremos estimar.

- Criando várias amostras para observar a distribuição amostral

```python
# Simular várias amostras e calcular a média de cada uma
n_amostras = 1000
tamanho_amostra = 20

medias = [np.mean(np.random.choice(dados, tamanho_amostra, replace=True))
          for _ in range(n_amostras)]

# Visualizar a distribuição das médias amostrais
sns.histplot(medias, kde=True)
plt.title('Distribuição amostral da média (largura da sépala)')
plt.xlabel('Média amostral')
plt.ylabel('Frequência')
plt.show()
```

- Explicação:

Pegamos 1000 amostras de tamanho 20 da população e calculamos a média de cada.
O gráfico mostra como a média amostral varia entre as amostras.
Repare que ela tem formato aproximadamente normal (isso é o Teorema Central do Limite).

- Mostrando a relação com o Teorema Central do Limite



Quando retiramos muitas amostras aleatórias de uma população (de qualquer forma), e calculamos a média de cada amostra, a distribuição dessas médias tende a se aproximar de uma distribuição normal, à medida que o tamanho da amostra aumenta, mesmo que a população original não seja normal.

```python
media_amostral = np.mean(medias)
desvio_amostral = np.std(medias)

print(f"Média das médias amostrais: {media_amostral:.2f}")
print(f"Desvio padrão das médias amostrais: {desvio_amostral:.3f}")
print(f"Erro padrão esperado: {desvio_pop / np.sqrt(tamanho_amostra):.3f}")
```

Explicação:

A média das médias amostrais deve ser muito próxima da média populacional,
e o desvio das médias (chamado erro padrão) se aproxima de 𝜎/𝑛.

Isso mostra que quanto maior o tamanho da amostra, menor a variabilidade das médias.

## **Estimação pontual e intervalar**

- Estimação pontual

É quando usamos um único valor da amostra (por exemplo, a média amostral) para estimar um parâmetro populacional.

- Estimação intervalar

É quando usamos um intervalo de confiança (IC) que contém o parâmetro com uma certa probabilidade.

```python
import scipy.stats as stats

# Estimar o IC de 95% para a média populacional
n = len(dados)
media = np.mean(dados)
sem = stats.sem(dados)
ic = stats.t.interval(0.95, df=n-1, loc=media, scale=sem)

print(f"Intervalo de confiança 95%: {ic}")
```

Usamos a distribuição t de Student porque o desvio padrão da população é desconhecido.

O IC de 95% significa que se repetíssemos o processo muitas vezes, 95% dos intervalos conteriam a média real da população.

```python
plt.figure(figsize=(6,4))
plt.errorbar(x=1, y=media, yerr=(ic[1]-ic[0])/2, fmt='o', color='red', capsize=5)
plt.axhline(media_pop, color='green', linestyle='--', label='Média populacional real')
plt.title('Intervalo de confiança da média')
plt.ylabel('Largura da sépala (cm)')
plt.legend()
plt.show()
```

O ponto vermelho é a média amostral e a barra mostra o intervalo de confiança.
A linha verde indica a média real da população — note que ela cai dentro do intervalo, o que é o esperado.

## **Teste de aderência**

O teste de aderência é usado para verificar se os dados seguem uma determinada distribuição teórica (como a normal, uniforme, etc.).

No Python, o **Teste de Kolmogorov-Smirnov (KS)** e o **Teste de Shapiro-Wilk** são frequentemente utilizados para verificar se os dados seguem uma distribuição normal.

1. **Teste de Shapiro-Wilk**:

* **O que faz**: Verifica se os dados seguem uma distribuição normal.

* **Hipóteses**:
  * H0: Os dados seguem uma distribuição normal.
  * H1: Os dados não seguem uma distribuição normal.

* **Interpretação**: Se o p-valor > 0.05, não rejeitamos H0 (os dados podem ser normais).


2. **Teste de Kolmogorov-Smirnov** (KS):

* **O que faz**: Compara a distribuição empírica dos dados com uma distribuição teórica (no caso, normal).

* **Hipóteses**:
  * H0: A distribuição dos dados é semelhante à distribuição teórica.
  * H1: A distribuição dos dados não é semelhante à distribuição teórica.

* **Interpretação**: Mesma regra do p-valor > 0.05.


* Definindo a Distribuição Teórica

No scipy.stats.kstest (ou outras implementações do teste KS), você informa explicitamente qual distribuição teórica quer comparar. Isso é feito pelo nome da distribuição e pelos seus parâmetros (se necessário).

Como especificar a distribuição:
  * Para normalidade: Usamos 'norm', que indica a distribuição normal padrão (média
𝜇 =0 e desvio padrão 𝜎 =1).
  * Para uniformidade: Usamos 'uniform', que indica a distribuição uniforme padrão no intervalo [0, 1].
  * Para outras distribuições: Passamos o nome da distribuição e seus parâmetros, se necessário, como 'expon' para distribuição exponencial, ou 'lognorm' para log-normal.



3. **Teste de D'Agostino-Pearson**:

* **O que faz**: Combina medidas de assimetria e curtose para verificar a normalidade dos dados.

* **Hipóteses**: Iguais aos testes anteriores.
* **Interpretação**: Idem.

```python
from scipy.stats import shapiro, kstest, normaltest

# Selecionar a variável para o teste (comprimento da pétala)
data = df['petal length (cm)']

# 1. Teste de Shapiro-Wilk
stat_shapiro, p_shapiro = shapiro(data)
print("Teste de Shapiro-Wilk:")
print(f"Estatística: {stat_shapiro:.4f}, p-valor: {p_shapiro:.4f}")
if p_shapiro > 0.05:
    print("Os dados seguem uma distribuição normal (não rejeita H0).\n")
else:
    print("Os dados não seguem uma distribuição normal (rejeita H0).\n")
```

```python
# 2. Teste de Kolmogorov-Smirnov (compara os dados a uma distribuição normal teórica)

from scipy.stats import norm

mean, std = data.mean(), data.std()  # Média e desvio padrão dos dados
stat_ks, p_ks = kstest(data, 'norm', args=(mean, std))
print("Teste de Kolmogorov-Smirnov:")
print(f"Estatística: {stat_ks:.4f}, p-valor: {p_ks:.4f}")
if p_ks > 0.05:
    print("Os dados seguem uma distribuição normal (não rejeita H0).\n")
else:
    print("Os dados não seguem uma distribuição normal (rejeita H0).\n")
```

```python
# 3. Teste de D'Agostino-Pearson

stat_dp, p_dp = normaltest(data)
print("Teste de D'Agostino-Pearson:")
print(f"Estatística: {stat_dp:.4f}, p-valor: {p_dp:.4f}")
if p_dp > 0.05:
    print("Os dados seguem uma distribuição normal (não rejeita H0).\n")
else:
    print("Os dados não seguem uma distribuição normal (rejeita H0).\n")
```

## **Teste de Hipótese**

Um teste de hipótese é um procedimento estatístico utilizado para avaliar se uma suposição sobre um parâmetro ou característica de uma população é plausível, com base em dados amostrais. Ele parte de duas hipóteses principais:

1. **Hipótese Nula (H₀)**: Representa a afirmação inicial ou padrão que se assume verdadeira até que se prove o contrário. Por exemplo, "não há diferença entre dois grupos" ou "o remédio não tem efeito".

2. **Hipótese Alternativa (H₁)**: Representa a afirmação contrária à hipótese nula. Por exemplo, "há diferença entre dois grupos" ou "o remédio tem efeito".

O teste avalia os dados e calcula a probabilidade de observar os resultados amostrais sob a suposição de que a hipótese nula seja verdadeira. Com base nessa probabilidade (p-valor), decidimos rejeitar H₀ ou não rejeitar H₀.

**Como usar o teste de hipótese no dia a dia?**


* **Definir o problema**:

 * **Exemplo**: Um fabricante quer saber se um novo método de produção reduz os custos em comparação ao método atual.


* **Formular as hipóteses**:

  * **Hipótese nula (H₀)**: O custo médio com o novo método é igual ao custo médio do método atual.
  * **Hipótese alternativa (H₁)**: O custo médio com o novo método é menor que o custo médio do método atual.


* **Coletar dados:**

Realize um experimento ou estudo observacional.

Exemplo: Meça os custos de produção com ambos os métodos.
Escolher o teste estatístico apropriado:

**Exemplos de testes**:

* t-test: Para comparar médias.
* chi-quadrado: Para variáveis categóricas.
* ANOVA: Para comparar mais de dois grupos.
* Teste Z: Para proporções.

**Calcular o p-valor e tomar decisões:**

Se o p-valor for menor que o nível de significância (α, geralmente 0,05): Rejeitamos H₀ e aceitamos H₁.

Se o p-valor for maior que α: Não rejeitamos H₀.


**Exemplo pratico**:

* Exemplo 1: **Teste t de Student**

* **Natureza**: Paramétrico

**Por quê** **?**:

* O Teste t de Student faz suposições sobre a distribuição dos dados:
A amostra deve vir de uma população que segue uma distribuição normal.

* A variância dentro dos grupos comparados deve ser homogênea (igual).

* Os dados devem ser quantitativos e em escala intervalar ou racional.

* Uso: Compara as médias de dois grupos para verificar diferenças significativas.

**Caso**:

Queremos verificar se a média do comprimento da pétala (petal length) da espécie setosa é diferente da média da espécie versicolor.


**Hipóteses**:

H₀: As médias do comprimento da pétala de setosa e versicolor são iguais.

H₁: As médias do comprimento da pétala de setosa e versicolor são diferentes.


```python
from scipy.stats import ttest_ind

# Filtrar as espécies
setosa = df[df['species'] == 'setosa']['petal length (cm)']
versicolor = df[df['species'] == 'versicolor']['petal length (cm)']

# Realizar o teste t
t_stat, p_value = ttest_ind(setosa, versicolor)

print("Estatística t:", t_stat)
print("Valor-p:", p_value)

# Avaliação do resultado
alpha = 0.05
if p_value < alpha:
    print("Rejeitamos H₀: Há diferença significativa nas médias.")
else:
    print("Não rejeitamos H₀: Não há diferença significativa nas médias.")
```

Exemplo 2: **Teste Qui-Quadrado**

Queremos verificar se há uma associação entre espécies e uma variável categórica. Neste caso, criaremos uma categorização do comprimento da pétala (petal length) em "curta" ou "longa".

**Natureza**: Paramétrico (em um sentido mais amplo, mas com ressalvas)

**Por quê?**:

* Baseia-se em aproximações estatísticas (distribuição qui-quadrado) para calcular o valor-p.

* É sensível ao tamanho da amostra e às frequências esperadas. Para amostras pequenas (frequências < 5), os resultados podem ser distorcidos.

* Uso: Testa associações entre variáveis categóricas em tabelas de contingência.
Exceção no uso do termo "paramétrico":

Embora frequentemente incluído entre testes paramétricos, o Teste Qui-Quadrado não depende de distribuições normais. Em vez disso, depende do tamanho e das frequências das amostras. Por isso, ele pode ser considerado semi-paramétrico.

**Caso**:

**Hipóteses**:

H₀: Não há associação entre a espécie e a categorização do comprimento da pétala.

H₁: Existe uma associação entre a espécie e a categorização do comprimento da pétala.


```python
from scipy.stats import chi2_contingency

# Criar uma coluna categorizando comprimento da pétala
df['petal_category'] = pd.cut(df['petal length (cm)'], bins=[0, 2, 7], labels=['curta', 'longa'])

# Criar tabela de contingência
contingency_table = pd.crosstab(df['species'], df['petal_category'])

# Realizar o teste Qui-Quadrado
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("Estatística Qui-Quadrado:", chi2_stat)
print("Valor-p:", p_value)

# Avaliação do resultado
alpha = 0.05
if p_value < alpha:
    print("Rejeitamos H₀: Existe associação entre a espécie e a categorização do comprimento da pétala.")
else:
    print("Não rejeitamos H₀: Não há associação entre a espécie e a categorização do comprimento da pétala.")
```

Quando temos frequências esperadas ou observadas menores que 5 em alguma célula da tabela de contingência, o Teste Exato de Fisher é mais apropriado do que o Teste Qui-Quadrado. Isso ocorre porque o Teste Qui-Quadrado depende de uma aproximação que não é confiável em amostras pequenas.

Exemplo 3: **Teste Exato de Fisher**

**Natureza**: Não Paramétrico

**Por quê?**:

* Não faz suposições sobre a distribuição dos dados.

* Calcula a probabilidade exata de observar uma tabela de contingência, dado o total de linhas e colunas.

* É especialmente útil para amostras pequenas ou quando as frequências esperadas são muito baixas.

* Uso: Testa a associação entre variáveis categóricas em tabelas 2x2 (ou maiores com extensões).

**Caso**:

H₀: Não há associação entre a espécie e a categorização do comprimento da pétala.

H₁: Existe uma associação entre a espécie e a categorização do comprimento da pétala.

```python
from scipy.stats import fisher_exact
import pandas as pd

# Criar uma coluna categorizando comprimento da pétala
df['petal_category'] = pd.cut(df['petal length (cm)'], bins=[0, 2, 7], labels=['curta', 'longa'])

# Criar tabela de contingência
contingency_table = pd.crosstab(df['species'], df['petal_category'])

# Exibir a tabela de contingência
print("Tabela de contingência:")
print(contingency_table)

# Transformar a tabela de contingência para analisar cada par de espécies
# Exemplo: Comparar 'setosa' vs 'versicolor'
setosa_versicolor = contingency_table.loc[['setosa', 'versicolor']].to_numpy()

# Verificar se a tabela é 2x2
if setosa_versicolor.shape == (2, 2):
    # Realizar o Teste Exato de Fisher
    oddsratio, p_value = fisher_exact(setosa_versicolor)

    print("\nResultado do Teste Exato de Fisher (Setosa vs Versicolor):")
    print("Valor-p:", p_value)

    # Avaliação do resultado
    alpha = 0.05
    if p_value < alpha:
        print("Rejeitamos H₀: Existe associação significativa entre a espécie e a categorização do comprimento da pétala.")
    else:
        print("Não rejeitamos H₀: Não há associação significativa entre a espécie e a categorização do comprimento da pétala.")
else:
    print("\nA tabela de contingência não é 2x2. O Teste Exato de Fisher não pode ser aplicado diretamente.")
```

**Obs**: O Teste Exato de Fisher exige tabelas 2x2. No exemplo acima, filtramos duas espécies específicas (setosa e versicolor). Isso cria uma tabela 2x2 extraída da tabela geral.

O teste foi diretamente aplicado usando a tabela 2x2. Se for necessário comparar mais pares de espécies, é só repetir o processo com outras combinações.

### Testes Paramétricos e Não Paramétricos

| **Objetivo do Teste**              | **Teste Paramétrico**        | **Teste Não Paramétrico**               |
|------------------------------------|------------------------------|------------------------------------------|
| Comparar a média de 2 grupos       | Teste t de Student           | Teste de Mann-Whitney (U de Mann-Whitney) |
| Comparar a média de 2 amostras emparelhadas | Teste t pareado              | Teste de Wilcoxon                        |
| Comparar a média de mais de 2 grupos | ANOVA                       | Teste de Kruskal-Wallis                  |
| Testar associações entre variáveis categóricas | Teste Qui-Quadrado         | Teste Exato de Fisher                    |
| Analisar correlação                | Correlação de Pearson        | Correlação de Spearman ou Kendall        |
| Testar normalidade                 | Teste de Shapiro-Wilk        | Teste de Kolmogorov-Smirnov              |


### **Testes Paramétricos**

**Definição**:

Os testes paramétricos assumem que os dados seguem uma distribuição específica, normalmente a distribuição normal (ou gaussiana). Além disso, eles requerem algumas condições sobre os dados, como a homogeneidade de variâncias e a escala de medição dos dados.

**Características**:

* Distribuição Normal: Assumem que a amostra vem de uma população com uma distribuição normal.

* Exigências sobre os dados:
Os dados precisam ser quantitativos.

* As variâncias entre os grupos ou amostras comparadas devem ser semelhantes (homocedasticidade).

**Quando Usar**:

Quando você tem grandes amostras e os dados seguem uma distribuição normal.
Quando as variáveis são quantitativas e a variância entre os grupos é homogênea.

### **Testes Não Paramétricos**

**Definição**:

Os testes não paramétricos, por outro lado, não fazem suposições sobre a distribuição dos dados. Eles podem ser usados em situações em que as condições dos testes paramétricos não são atendidas, como em dados não normais, qualitativos, ou com amostras pequenas.



**Características**:

* Sem suposição de distribuição: Não há a suposição de que os dados sigam uma distribuição específica.

* Mais flexíveis: Podem ser usados para dados qualitativos ou quantitativos que não atendem à normalidade.


**Quando Usar**:
* Quando as amostras são pequenas e não seguem uma distribuição normal.
* Quando os dados são qualitativos (categóricos) ou ordinal.
* Quando as condições de homogeneidade de variâncias não são atendidas.

### Comparação entre Testes Paramétricos e Não Paramétricos

| **Característica**                   | **Testes Paramétricos**                   | **Testes Não Paramétricos**               |
|--------------------------------------|------------------------------------------|------------------------------------------|
| **Distribuição dos dados**           | Assume distribuição normal               | Não assume nenhuma distribuição específica|
| **Tipo de dados**                    | Dados quantitativos (intervalares ou racionais) | Dados qualitativos ou ordinais           |
| **Exemplo de Testes**                | Teste t de Student, ANOVA, Correlação de Pearson | Mann-Whitney, Kruskal-Wallis, Spearman   |
| **Requisitos**                       | Normalidade, homogeneidade de variâncias | Não exige normalidade ou homogeneidade   |
| **Aplicabilidade**                   | Quando os dados são normais e grandes amostras | Quando os dados são não normais ou amostras pequenas |
